```python
from collections import deque

def bfs(matrix, queue, n, m, visited):
  if len(queue) == 0:
    print("No path found")
    return

  current_path = queue.popleft()
  current = current_path[-1]  # Last cell in path is current position
  x, y = current[0], current[1]
  print(current)

  # Check if already visited
  if current in visited:
    return bfs(matrix, queue, n, m, visited)

  # Check if wall
  if matrix[x][y] == '1':
    visited.append(current)
    bfs(matrix, queue, n, m, visited)
    return

  # Check if Goal
  if matrix[x][y] == 'G':
    print("Path found:", current_path)
    return

  # Mark as visited
  visited.append(current)

  # Add neighbors with full path
  if x != 0:
    queue.append(current_path + [(x-1, y)])
  if x != n-1:
    queue.append(current_path + [(x+1, y)])
  if y != 0:
    queue.append(current_path + [(x, y-1)])
  if y != m-1:
    queue.append(current_path + [(x, y+1)])

  bfs(matrix, queue, n, m, visited)


matrix = list((['S', '0', '1'],
              ['0', '1', '0'],
              ['0', 'G', '1']))

queue = deque()
queue.append([(0, 0)])   # Queue now stores full paths instead of single cells
visited = list()

bfs(matrix, queue, 3, 3, visited)
```
