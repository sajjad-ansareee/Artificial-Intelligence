```python
# Q.6
import numpy as np
arr = np.random.randint(-50, 151, size=100)
arr_reshaped = arr.reshape((10, 10))
arr_reshaped[arr_reshaped < 0] = 0
arr_min = arr_reshaped.min()
arr_max = arr_reshaped.max()
normalized = (arr_reshaped - arr_min) / (arr_max - arr_min)
row_means = normalized.mean(axis=1)
print("Original reshaped array (negatives replaced):\n", arr_reshaped)
print("\nNormalized array:\n", normalized)
print("\nRow-wise means:\n", row_means)


'''
2D array of 10*10 size with negative values assigned as 0
Normalized the array
'''
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols_to_plot = ['Glucose', 'BMI', 'Insulin', 'Age']

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten()

# Iterate through the columns and create box plots
for i, col in enumerate(cols_to_plot):
    sns.boxplot(x='Outcome', y=col, data=file, ax=axes[i], palette='Set2')
    axes[i].set_title(f'Distribution of {col} by Outcome', fontsize=14)
    axes[i].set_xlabel('Outcome (0: Non-Diabetic, 1: Diabetic)', fontsize=12)
    axes[i].set_ylabel(col, fontsize=12)

plt.tight_layout()
```
