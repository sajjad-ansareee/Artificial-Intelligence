```python
letters, spaces, upper, lower, digits = 0, 0, 0, 0, 0
with open("/myfile.txt", "r") as file:
  content=file.read()
  for i in content:
    if i.isalpha():
      letters+=1
    if i==' ':
      spaces+=1
    if i.isupper():
      upper+=1
    if i.islower():
      lower+=1
    if i.isdigit():
      digits+=1
  print (f"Letters: {letters}")
  print (f"Spaces: {spaces}")
  print (f"Upper: {upper}")
  print (f"Lower: {lower}")
  print (f"digits: {digits}")
```

```python
from collections import deque

MOVES = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # Right, Down, Left, Up


def all_paths_dfs(rows, cols, goal):
    all_paths = []

    def dfs(x, y, path, visited):
        if (x, y) == goal:
            all_paths.append(path.copy())
            return

        for dx, dy in MOVES:
            nx, ny = x + dx, y + dy
            if 0 <= nx < rows and 0 <= ny < cols and (nx, ny) not in visited:
                visited.add((nx, ny))
                path.append((nx, ny))
                dfs(nx, ny, path, visited)
                path.pop()
                visited.remove((nx, ny))

    dfs(0, 0, [(0, 0)], {(0, 0)})
    return all_paths


def shortest_path_bfs(rows, cols, goal):
    queue = deque([((0, 0), [(0, 0)])])
    visited = {(0, 0)}

    while queue:
        (x, y), path = queue.popleft()
        if (x, y) == goal:
            return path

        for dx, dy in MOVES:
            nx, ny = x + dx, y + dy
            if 0 <= nx < rows and 0 <= ny < cols and (nx, ny) not in visited:
                visited.add((nx, ny))
                queue.append(((nx, ny), path + [(nx, ny)]))

    return None


rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))
gx = int(input("Enter goal row: "))
gy = int(input("Enter goal column: "))

goal = (gx, gy)

all_paths = all_paths_dfs(rows, cols, goal)
shortest = shortest_path_bfs(rows, cols, goal)

print("\nTotal possible paths:", len(all_paths))
print("\nAll paths:")
for p in all_paths:
    print(p)

print("\nShortest path:")
print(shortest)
print("Shortest path length:", len(shortest) - 1)
```
