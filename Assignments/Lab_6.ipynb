```python
from random import randint
# firstly I am considering it for the rook
def initialize(board_size):
    board=[0 for _ in range(board_size)]
    for _ in range(board_size):
        board[_]=randint(0, board_size-1)
    return board
```

```python
# there will be a list of children, and we have to return the best one
import math
def next_better(children):
    # the worst case in which each rook attacking other
    best_one=list()
    minimum=math.comb(len(children), 2)
    for i in range(len(children)):
        current_fitness=fitness_function(children[i])
        if current_fitness < minimum:
            best_one=children[i].copy()
            minimum=current_fitness
    return [best_one, minimum]
```

```python
# randomly change any of the rook
def get_children(board):
    children=list()
    for i in range(len(board)):
        child=board.copy()
        child[i]=randint(0, len(board)-1)
        children.append(child)
    return children
```

```python
def hill_climb(board):
    current_fitness=fitness_function(board)
    if current_fitness==0:
        print('The found solution: ', board)
        return
    children=get_children(board)
    print('Children:', children)
    next_board, next_fitness=next_better(children)
    print('Next better:', next_board, next_fitness)
    if next_fitness<current_fitness:
        hill_climb(next_board)
    # if we are stuck like there is no better child, so we will have local optima
    else:
        print('The local optima: ', board)
```
