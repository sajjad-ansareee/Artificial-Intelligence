```python
from random import randint
# firstly I am considering it for the rook
def initialize(board_size):
    board=[0 for _ in range(board_size)]
    for _ in range(board_size):
        board[_]=randint(0, board_size-1)
    return board
```

```python
# count how many same numbers in the row
from math import comb
def fitness_function(board):
    count=0
    size=len(board)
    for i in range(size):
        for j in range(i, size):
            if board[i]==board[j] and i!=j:
                count+=1
            # now also checking for the diagonal for queen move
            elif board[i]==(board[j]+j-1) and i!=j:
                count+=1
            elif board[i]==(board[j]-j+i) and i!=j:
                count+=1
    return comb(len(board), 2)-count
```

```python
def crossover(mom, dad, point):
    return [(mom[:point]+dad[point:]),
            (dad[:point]+mom[point:])]
```

```python
# of all the given children we will randomly mutate ratio child
def mutate(offsprings, ratio):
    # firstly calculate how many offsprings do we have
    # find length of each of the child
    # then randomly mutate them
    total=len(offsprings)*len(offsprings[0])
    for _ in range(int(round(total*ratio)/2)):
        i, j=randint(0, len(offsprings)-1), randint(0, len(offsprings[0])-1)
        a, b=randint(0, len(offsprings)-1), randint(0, len(offsprings[0])-1)
        offsprings[i][j], offsprings[a][b]=offsprings[a][b], offsprings[i][j]
    return offsprings

```

```python
import matplotlib.pyplot as plt
def plot_comparison(list1, list2):
    plt.plot(list1, marker='o', label='Best fitness')
    plt.plot(list2, marker='o', label='Average fitness')
    plt.title('Genetic Algorithm')
    plt.grid()
    plt.legend()
    plt.show()
# plot_comparison(a, b)
```

```python
def get_fitness_values(boards):
    fitness_values=[0 for _ in range(len(boards))]
    for _ in range(len(boards)):
        fitness_values[_]=(fitness_function(boards[_]))
    return fitness_values
```

```python
def get_offsprings(boards):
    offsprings=[[] for _ in range(10)]
    crossover_point=2
    for _ in range(5):
        offsprings[_*2], offsprings[(_*2)+1]=crossover(boards[_*2], boards[(_*2)+1], crossover_point)
    # now mutate the offsprings
    mutation_ratio=0.5
    offsprings=mutate(offsprings, mutation_ratio)
    return offsprings
```

```python
def current_boards_information(boards, fitness_values):
    probabilities=[0 for _ in range(10)]
    estimated_count=[0 for _ in range(10)]
    actual_count=[0 for _ in range(10)]
    for _ in range(len(boards)):
        probabilities[_]=round((fitness_values[_]/sum(fitness_values)), 2)
        estimated_count[_]=round(probabilities[_]*10, 2)
        actual_count[_]=int(round(estimated_count[_]))
        print(boards[_], fitness_values[_], probabilities[_], estimated_count[_], actual_count[_])
```

```python
def main():
    rows, columns=10, 10
    fitness_values=list()
    best_fitness=list()
    average_fitness=list()
    boards=[[] for _ in range(rows)]
    for _ in range(rows):
        boards[_]=initialize(columns)
    # simply taking all the children instead of considering probability
    # now generate new offsprings from parents
    loop_count=30
    while loop_count:
        loop_count-=1
        fitness_values=get_fitness_values(boards)
        current_boards_information(boards, fitness_values)
        print(f'Best fitness: {max(fitness_values)}, Average fitness: {sum(fitness_values)/len(fitness_values)}')
        best_fitness.append(max(fitness_values))
        average_fitness.append(sum(fitness_values)/len(fitness_values))
        boards=get_offsprings(boards)
    plot_comparison(best_fitness, average_fitness)
    print(f'The placement of board after algorithm: {boards[fitness_values.index(max(fitness_values))]}')
main()
```
